```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import IPython.display as ipd
import librosa
import librosa.display
from itertools import cycle
```

```python
filename = (r"C:\Users\pothi\Downloads\speech audio.wav")
Audio(filename =filename)
```

```python
y, sr = librosa.load(filename)
librosa.display.waveshow(y)
plt.title('Speech Signal Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
## Total no of samples and range of the amplitude and the spoken words are with high amplitude and silent are close to zero
print(f"Signal length: {len(y)} samples")
print(f"Magnitude range: {min(y)} to {max(y)}")
```

```python

start_time = 1 
end_time = 2  

segment = y[int(start_time * sr):int(end_time * sr)]

# Playing  the audio segment
ipd.Audio(segment, rate=sr)
```

```python
plot(figure(10,4))
librosa.display.waveshow(segment, sr=sr)
plt.title(f'Waveform of Audio Segment ({start_time} to {end_time} seconds)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
reversed_y = np.flip(y)

plt.figure(figsize=(10, 4))
librosa.display.waveshow(reversed_y, sr=sr)
plt.title('Waveform of Reversed Audio')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```
